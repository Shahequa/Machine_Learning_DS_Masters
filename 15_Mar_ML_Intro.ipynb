### Question 2: What is supervised learning? List some examples of supervised learning.

**Ans)**
`Supervised learning is a type of machine learning technique in which an algorithm is trained to make predictions or decisions based on labeled data. The labeled data consists of input features and corresponding output labels or target values, which are used to teach the algorithm how to make accurate predictions on new, unseen data.`

*Examples of supervised learning include:*

* Classification: predicting whether an input belongs to a specific category or class. For example, classifying emails as spam or not spam, identifying the breed of a dog based on an image, or predicting whether a customer is likely to buy a product or not.

* Regression: predicting a continuous value output based on input features. For example, predicting house prices based on features like location, square footage, and number of bedrooms, or predicting the amount of rainfall based on temperature and humidity data.

* Object detection: identifying the location and type of objects in an image or video. For example, identifying pedestrians, cars, and traffic signs in self-driving car applications.

* Natural language processing (NLP): analyzing and processing human language data. For example, sentiment analysis to predict whether a review is positive or negative, or text classification to categorize news articles into different topics.
    
    *Supervised learning is widely used in many real-world applications, such as predicting customer behavior, medical diagnosis, image and speech recognition, and many others. The key advantage of supervised learning is that it can leverage labeled data to learn complex patterns and make accurate predictions on new, unseen data.*

### Question 3 : What is unsupervised learning? List some examples of unsupervised learning.

**Ans)**
`Unsupervised learning is a type of machine learning technique in which an algorithm learns to identify patterns and relationships in data without explicit guidance or labeled examples. Unlike supervised learning, the data in unsupervised learning is not labeled with the correct output values or target labels. The goal of unsupervised learning is to discover hidden structures, clusters, or relationships within the data.`

*Examples of unsupervised learning include:*

* Clustering: grouping similar data points together based on their inherent similarities. For example, clustering customer data to identify segments of customers with similar behaviors and preferences, or clustering genetic data to identify patterns of gene expression.

* Dimensionality reduction: reducing the number of features in the data while preserving as much of the original information as possible. For example, using principal component analysis (PCA) to reduce the dimensionality of high-dimensional data such as images or audio signals, or using t-SNE to visualize high-dimensional data in a lower-dimensional space.

* Anomaly detection: identifying unusual or abnormal data points that do not conform to the normal patterns in the data. For example, detecting credit card fraud based on unusual transaction patterns or identifying network intrusions based on anomalous traffic patterns.
    
* Association rule mining: discovering interesting associations or correlations between different items or variables in the data. For example, identifying frequent itemsets in market basket data to recommend related products or discovering co-occurrence patterns in user behavior data to personalize recommendations.

   *In unsupervised learning, the algorithm has to find structure on its own without any labels. This makes unsupervised learning an exploratory technique that can uncover hidden patterns in data that may not be immediately apparent. Unsupervised learning can be useful in many real-world applications, such as customer segmentation, fraud detection, and anomaly detection.*

### Question 5 : What are main differences between supervised, unsupervised and semi-supervised learning?

**Ans)**
`The main differences between supervised, unsupervised, and semi-supervised learning are:`

    Supervised learning: In supervised learning, the algorithm is trained on labeled data, which means the input data is accompanied by corresponding output labels or target values. The goal is to learn a mapping from input features to output labels or target values, and then make predictions on new, unseen data. Supervised learning is used for tasks such as classification, regression, and object detection.

    Unsupervised learning: In unsupervised learning, the algorithm is trained on unlabeled data, which means there are no output labels or target values. The goal is to discover hidden patterns, structures, or relationships in the data without any prior knowledge. Unsupervised learning is used for tasks such as clustering, dimensionality reduction, and anomaly detection.

    Semi-supervised learning: In semi-supervised learning, the algorithm is trained on a combination of labeled and unlabeled data. The goal is to leverage the labeled data to improve the accuracy of the model on unlabeled data. Semi-supervised learning is particularly useful when labeled data is scarce or expensive to obtain. Examples of semi-supervised learning include co-training, self-training, and multi-view learning.

  *In summary, supervised learning requires labeled data, unsupervised learning does not require labeled data, and semi-supervised learning combines both labeled and unlabeled data. The choice of which type of learning to use depends on the specific task and the availability of labeled data.*

### Question 6 : What is train, test and validation split? Explain Importance of each term.

**Ans)**
`Train, test, and validation split is a common practice in machine learning to evaluate the performance of a model on new, unseen data. The dataset is divided into three subsets:`

    Train set: The training set is used to train the model. It contains labeled data that is used to learn the patterns and relationships between input features and output labels or target values.

    Test set: The test set is used to evaluate the performance of the trained model on new, unseen data. It contains labeled data that the model has not seen during training.

    Validation set: The validation set is used to tune the hyperparameters of the model. Hyperparameters are parameters that are set before training, such as the learning rate or the number of hidden layers. The validation set is used to select the best hyperparameters that give the highest accuracy on the test set.

`The importance of each split is as follows:`

    Train set: The train set is important because it is used to train the model. The model learns the patterns and relationships in the data from the training set, and this knowledge is used to make predictions on new, unseen data.
    
    Test set: The test set is important because it is used to evaluate the performance of the trained model on new, unseen data. The test set is used to estimate how well the model will perform on new data in the real world.

    Validation set: The validation set is important because it is used to tune the hyperparameters of the model. The hyperparameters control how the model is trained and how it learns from the data. The validation set is used to select the best hyperparameters that give the highest accuracy on the test set.

  *In summary, train, test, and validation split is an important technique in machine learning to evaluate the performance of a model on new, unseen data. The train set is used to train the model, the test set is used to evaluate the performance, and the validation set is used to tune the hyperparameters. The importance of each split is to ensure that the model is accurate, generalizes well, and can be deployed in the real world.*

### Question 7 : How can unsupervised machine learning be used in anomaly detection ?

**Ans)**
`Unsupervised machine learning can be used in anomaly detection by identifying data points that do not conform to the normal patterns or structures in the data. Anomalies are data points that are rare or unusual, and they can indicate errors, fraud, or other interesting events in the data.`

`Here are the steps for using unsupervised machine learning in anomaly detection:`

* Data preprocessing: The first step is to preprocess the data, which includes cleaning, transforming, and normalizing the data. This step is important to ensure that the data is in a suitable format for unsupervised machine learning algorithms.

* Feature extraction: The second step is to extract features from the data, which means selecting the most relevant and informative variables that describe the data. Feature extraction is important because it reduces the dimensionality of the data and focuses on the most important aspects of the data.

* Unsupervised learning: The third step is to apply unsupervised machine learning algorithms to the data, such as clustering, dimensionality reduction, or density estimation. These algorithms learn the patterns and structures in the data without any prior knowledge of the anomalies.
    
* Anomaly detection: The fourth step is to detect anomalies in the data by identifying data points that deviate significantly from the normal patterns or structures. Anomalies can be detected by measuring the distance, density, or similarity of the data points to the normal patterns or structures. Some common methods for anomaly detection include isolation forest, one-class SVM, and local outlier factor.

* Evaluation: The final step is to evaluate the performance of the unsupervised machine learning algorithm on detecting anomalies. This can be done by comparing the detected anomalies to a ground truth or by measuring the precision, recall, or F1-score of the algorithm.
